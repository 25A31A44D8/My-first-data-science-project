```python
import matplotlib.pyplot as plt

resumes = [
    "Python Data Analysis Machine Learning",
    "Java SQL Backend Development",
    "Python Deep Learning Data Science"
]

job_description = "Looking for Python Data Science candidate"

results = []

for resume in resumes:
    count = 0
    for word in job_description.split():
        if word.lower() in resume.lower():
            count += 1
    results.append((resume, count))

results.sort(key=lambda x: x[1], reverse=True)

resume_names = []
scores = []

for i in range(len(results)):
    resume_names.append("Resume " + str(i+1))
    scores.append(results[i][1])

plt.bar(resume_names, scores)
plt.xlabel("Resumes")
plt.ylabel("Match Score")
plt.title("Resume Matching Scores")
plt.show()
```
